```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_image = cv2.imread("Rose 1024x1024.tif", cv2.IMREAD_GRAYSCALE)
original_image = cv2.resize(original_image, (512, 512))    
```

```python
def decrease_resolution(image):
    height, width = image.shape
    decreased_image = np.zeros((height //2, width // 2))
    
    for r in range(0, height, 2):
        for c in range(0, width, 2):
            decreased_image[r //2, c // 2 ] = image[r, c]
    return np.uint8(decreased_image)
        

    
```

```python
decreaased_image = original_image.copy()
plt.figure(figsize=(10, 10))

for k in range(1, 5):
    decreaased_image = decrease_resolution(decreaased_image)
    plt.subplot(2, 2, k)
    plt.imshow(decreaased_image, cmap='gray')
    height, width = decreaased_image.shape
    plt.title(f"{height} x {width}")

plt.show()
```

```python
original_image = cv2.imread("Skull 374x452.tif", cv2.IMREAD_GRAYSCALE)
```

```python
def decrease_intensity(image, number_of_bits):
    step = 255 / (2 ** number_of_bits - 1)
    height, width = image.shape
    decreased_image = image.copy()
    
    for r in range(height):
        for c in range(width):
            decreased_image[r, c] = np.round(image[r, c] / step) * step
    return np.uint8(decreased_image)
```

```python
decreaased_image = original_image.copy()
plt.figure(figsize = (13, 8))


for k in range(1, 9):
    number_of_bits = 9 - k
    decreaased_image = decrease_intensity(decreaased_image, number_of_bits)
    plt.subplot(2, 4, k)
    plt.imshow(decreaased_image, cmap='gray')
    height, width = decreaased_image.shape
    plt.title(f"{number_of_bits}-Bits Image")


plt.show()
```
